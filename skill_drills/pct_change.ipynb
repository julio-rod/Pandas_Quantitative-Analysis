```python
import pandas as pd
from pathlib import Path
```

```python
#Review data
sp500_csv = pd.read_csv(
    Path("./Resources/sp500_close.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True)
sp500_csv.head()
```

```python
#Calculate the daily return using the `pct_change()` function
daily_returns = sp500_csv.pct_change() #By using the pct_change function, 
# we can generate the daily return calculations with one simple line of code. Financial analysts, rejoice!
```

```python
daily_returns.head() 
```

```python
#using the plot function 
daily_returns.plot(figsize=(10,7), title="S&P 500 Daily Returns May 2014-2019", color="green")
```

```python
# Calculating daily return for a 1 year period
daily_returns_1_year = daily_returns.loc["2018-05-01":"2019-05-01"]
daily_returns_1_year.describe()
```

```python
#using the plot function 
daily_returns_1_year.plot(figsize=(10,7), title="S&P 500 Daily Returns May 2018-2019", color="pink")
```

```python
# Calculating daily return for a 3 year period
daily_returns_3_year = daily_returns.loc["2016-05-01":"2019-05-01"]
daily_returns_3_year.describe()
```

Whether we use shift(1) or pct_change to calculate the daily return data, the describe function gives us an initial understanding of that data. 
We can then enhance our understanding through Pandas visualization functions, like plot. -SPOT

```python
#using the plot function 
daily_returns_3_year.plot(figsize=(10,7), title="S&P 500 Daily Returns May 2016-2019", color="orange")
```

```python
# Calculating daily return for a 5 year period
daily_returns_5_year = daily_returns.loc["2014-05-01":"2019-05-01"]
daily_returns_5_year.describe()
```

```python
#using the plot function 
daily_returns_5_year.plot(figsize=(10,7), title="S&P 500 Daily Returns May 2014-2019")
```
